```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv',sep=',')
#display(df)
```

```python
plt.scatter(df['Year'],df['uAMHz'])
plt.ylabel("CPU Power (uA/MHz)",fontsize=18,labelpad=10)
plt.xlabel("Year",fontsize=18,labelpad=10)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
plt.scatter(df['Year'],df['SRAM'])
plt.ylabel("SRAM (kB)",fontsize=18,labelpad=10)
plt.xlabel("Year",fontsize=18,labelpad=10)
plt.hlines(144,1990,2021,color='red')
plt.xlim((1995,2021))
plt.text(2021.5,130,"PDP 11/45",fontsize=16,color='red')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
plt.scatter(df['Year'],df['Flash'])
plt.ylabel("Flash (kB)",fontsize=18,labelpad=10)
plt.xlabel("Year",fontsize=18,labelpad=10)
plt.hlines(1000,1990,2021,color='red')
plt.text(2021.5,950,"PDP 11/45",fontsize=16,color='red')
plt.xlim((1995,2021))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```
